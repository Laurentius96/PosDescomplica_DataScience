## Aula 01: Probabilidade - Variável Aleatória

<br>

### Introdução

Nesta aula, exploramos o conceito de variáveis aleatórias, que são fundamentais na teoria da probabilidade. Uma variável aleatória é uma função que associa um número real a cada resultado possível de um experimento aleatório.

### Conceitos Principais

- **Variável Aleatória Discreta:** Assume um conjunto finito ou enumerável de valores. Exemplos: número de caras em lançamentos de moedas.
  
- **Variável Aleatória Contínua:** Pode assumir qualquer valor dentro de um intervalo contínuo. Exemplos: altura de uma pessoa, tempo de espera.

### Exemplo de Aplicação

```python
import numpy as np

# Exemplo de variável aleatória discreta
lancamentos = np.random.binomial(n=10, p=0.5, size=1000)
print(f"Média dos lançamentos: {np.mean(lancamentos)}")
```

## Aula 02: Teorema das Probabilidades

<br>

### Introdução

O Teorema das Probabilidades é uma regra fundamental para o cálculo de probabilidades, incluindo a probabilidade condicional e a regra da soma e produto.

### Conceitos Principais

- **Regra da Soma:** Se A e B são eventos mutuamente exclusivos, a probabilidade de A ou B ocorrer é a soma das probabilidades de A e B.
  
- **Regra do Produto:** A probabilidade de dois eventos independentes ocorrerem juntos é o produto de suas probabilidades individuais.

### Exemplo de Aplicação

```python
# Exemplo de regra do produto
prob_a = 0.3
prob_b = 0.4
prob_a_e_b = prob_a * prob_b
print(f"Probabilidade de A e B ocorrerem: {prob_a_e_b}")
```

## Aula 03: Probabilidade Condicional

<br>

### Introdução

A probabilidade condicional é a probabilidade de um evento ocorrer dado que outro evento já ocorreu. É essencial para entender como eventos dependem uns dos outros.

### Conceitos Principais

- **Probabilidade Condicional:** \( P(A|B) = \frac{P(A \cap B)}{P(B)} \) se \( P(B) > 0 \).

- **Independência:** Dois eventos A e B são independentes se \( P(A|B) = P(A) \).

### Exemplo de Aplicação

```python
# Exemplo de probabilidade condicional
def prob_condicional(prob_a_e_b, prob_b):
    return prob_a_e_b / prob_b

prob_a_e_b = 0.2
prob_b = 0.5
prob_a_dado_b = prob_condicional(prob_a_e_b, prob_b)
print(f"Probabilidade de A dado B: {prob_a_dado_b}")
```

## Aula 04: Teorema de Bayes

<br>

### Introdução

O Teorema de Bayes é uma ferramenta poderosa para atualizar a probabilidade de uma hipótese com base em novas evidências. Ele é amplamente utilizado em aprendizado de máquina e inferência estatística.

### Conceitos Principais

- **Teorema de Bayes:** \( P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \)

### Exemplo de Aplicação

```python
# Exemplo do Teorema de Bayes
prob_a = 0.01  # Probabilidade a priori
prob_b_dado_a = 0.8  # Sensibilidade
prob_b = 0.05  # Probabilidade marginal
prob_a_dado_b = (prob_b_dado_a * prob_a) / prob_b
print(f"Probabilidade a posteriori de A dado B: {prob_a_dado_b}")
```
