## Aula 01: Probabilidade - Variável Aleatória

### Introdução

Nesta aula, exploramos o conceito de variáveis aleatórias, fundamentais para a teoria da probabilidade. Uma variável aleatória é uma função que associa um número real a cada resultado possível de um experimento aleatório. Elas são essenciais para modelar incertezas em diversos cenários do mundo real.

### Conceitos Principais

- **Variável Aleatória Discreta:** Assume um conjunto finito ou enumerável de valores.
  - **Exemplo:** Número de caras em lançamentos de moedas.
  
- **Variável Aleatória Contínua:** Pode assumir qualquer valor dentro de um intervalo contínuo.
  - **Exemplo:** Altura de uma pessoa, tempo de espera.

### Exemplo de Aplicação

No mercado de trabalho, variáveis aleatórias são amplamente utilizadas em análises de risco, modelagem de tempo de espera em filas, e na simulação de cenários financeiros.

**Exemplo de Variável Aleatória Discreta:**
```python
import numpy as np

# Exemplo de variável aleatória discreta: número de caras em 10 lançamentos de uma moeda
lancamentos = np.random.binomial(n=10, p=0.5, size=1000)
print(f"Média dos lançamentos: {np.mean(lancamentos)}")
```

**Exemplo de Variável Aleatória Contínua:**
```python
import numpy as np

# Exemplo de variável aleatória contínua: altura de pessoas (em metros) seguindo uma distribuição normal
alturas = np.random.normal(loc=1.75, scale=0.1, size=1000)
print(f"Média das alturas: {np.mean(alturas)}")
```

## Aula 02: Teorema das Probabilidades

### Introdução

O Teorema das Probabilidades é uma regra fundamental para o cálculo de probabilidades, incluindo a probabilidade condicional e as regras da soma e produto. Ele é aplicado para modelar a ocorrência de eventos em diversos campos, desde finanças até engenharia.

### Conceitos Principais

- **Regra da Soma:** Se A e B são eventos mutuamente exclusivos, a probabilidade de A ou B ocorrer é a soma das probabilidades de A e B.
  - **Exemplo:** Calcular a probabilidade de falha em um sistema onde dois componentes distintos podem falhar.
  
- **Regra do Produto:** A probabilidade de dois eventos independentes ocorrerem juntos é o produto de suas probabilidades individuais.
  - **Exemplo:** Probabilidade de sucesso em duas etapas independentes de um projeto.

### Exemplo de Aplicação

No contexto empresarial, essas regras ajudam a calcular a probabilidade combinada de eventos como lançamentos de produtos e campanhas de marketing.

**Exemplo de Regra da Soma:**
```python
# Exemplo de regra da soma: probabilidade de falha em dois componentes mutuamente exclusivos
prob_componente_1 = 0.1
prob_componente_2 = 0.2
prob_falha_total = prob_componente_1 + prob_componente_2
print(f"Probabilidade de falha em pelo menos um componente: {prob_falha_total}")
```

**Exemplo de Regra do Produto:**
```python
# Exemplo de regra do produto: probabilidade de sucesso em duas etapas independentes de um projeto
prob_etapa_1 = 0.8
prob_etapa_2 = 0.7
prob_sucesso_total = prob_etapa_1 * prob_etapa_2
print(f"Probabilidade de sucesso em ambas as etapas: {prob_sucesso_total}")
```

## Aula 03: Probabilidade Condicional

### Introdução

A probabilidade condicional é a probabilidade de um evento ocorrer dado que outro evento já ocorreu. Ela é crucial para entender como eventos dependem uns dos outros, sendo amplamente utilizada em modelos preditivos e análises de comportamento.

### Conceitos Principais

- **Probabilidade Condicional:** image.png
  - **Exemplo:** Probabilidade de um cliente comprar um produto dado que ele já adicionou outro produto ao carrinho.
  
- **Independência:** Dois eventos A e B são independentes se image.png
  - **Exemplo:** Lançamentos sucessivos de uma moeda, onde o resultado de um lançamento não afeta o outro.

### Exemplo de Aplicação

Probabilidade condicional é usada em diversas áreas, como marketing (para segmentação de clientes), finanças (análise de portfólios) e engenharia (análise de falhas).

**Exemplo de Probabilidade Condicional:**
```python
# Exemplo de probabilidade condicional: compra de um produto dado que outro já foi adicionado ao carrinho
prob_produto_b_no_carrinho = 0.2
prob_produto_a_e_b = 0.1
prob_produto_a_dado_b = prob_produto_a_e_b / prob_produto_b_no_carrinho
print(f"Probabilidade de o cliente comprar o produto A dado que já adicionou o produto B ao carrinho: {prob_produto_a_dado_b}")
```

**Exemplo de Independência:**
```python
# Exemplo de independência: lançamentos sucessivos de uma moeda
lancamento_1 = np.random.choice(['cara', 'coroa'], p=[0.5, 0.5], size=1000)
lancamento_2 = np.random.choice(['cara', 'coroa'], p=[0.5, 0.5], size=1000)
prob_cara_dado_cara = np.mean(lancamento_1 == 'cara') * np.mean(lancamento_2 == 'cara')
print(f"Probabilidade de obter cara em ambos os lançamentos: {prob_cara_dado_cara}")
```

## Aula 04: Teorema de Bayes

### Introdução

O Teorema de Bayes é uma ferramenta poderosa para atualizar a probabilidade de uma hipótese com base em novas evidências. Ele é amplamente utilizado em aprendizado de máquina, diagnóstico médico e inferência estatística.

### Conceitos Principais

- **Teorema de Bayes:** image.png
  - **Exemplo:** Atualizar a probabilidade de um paciente ter uma doença com base no resultado de um teste diagnóstico.

### Exemplo de Aplicação

O Teorema de Bayes é amplamente utilizado em várias indústrias, desde a medicina (para diagnósticos) até finanças (para análise de risco).

**Exemplo de Teorema de Bayes:**
```python
# Exemplo do Teorema de Bayes: cálculo da probabilidade de um paciente ter uma doença após um teste positivo
prob_doenca = 0.01  # Probabilidade a priori
prob_teste_positivo_dado_doenca = 0.8  # Sensibilidade do teste
prob_teste_positivo = 0.05  # Probabilidade de teste positivo
prob_doenca_dado_teste_positivo = (prob_teste_positivo_dado_doenca * prob_doenca) / prob_teste_positivo
print(f"Probabilidade de o paciente ter a doença dado que o teste foi positivo: {prob_doenca_dado_teste_positivo}")
```
